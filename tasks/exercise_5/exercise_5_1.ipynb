```python
from geometry_msgs.msg import PoseStamped
import rospy
import sys
from moveit_commander import MoveGroupCommander, roscpp_initialize, roscpp_shutdown, PlanningSceneInterface

"""
Ziel :
Ein Objekt zur Szene hinzufügen, was der Roboterarm dann greifen kann.
"""

# ROS und MoveIt! initialisieren
roscpp_initialize(sys.argv)
rospy.init_node('add_object', anonymous=True)

# Interface für die Planungsumgebung
scene = PlanningSceneInterface()
rospy.sleep(1)

# Objekt zur Szene hinzufügen
box_pose = PoseStamped()
box_pose.header.frame_id = "panda_link0"  # Basis-Frame des Roboters
box_pose.pose.position.x = 0.4  # X-Position des Objekts
box_pose.pose.position.y = 0.0  # Y-Position des Objekts
box_pose.pose.position.z = 0.1  # Z-Position des Objekts (Tischhöhe)
box_pose.pose.orientation.w = 1.0

scene.add_box("target_object", box_pose, size=(0.05, 0.05, 0.05))  # Würfel mit 5x5x5 cm
rospy.loginfo("Objekt zur Szene hinzugefügt.")
rospy.sleep(2)
```
