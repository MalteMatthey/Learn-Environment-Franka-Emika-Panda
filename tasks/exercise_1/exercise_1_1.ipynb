First, we load the required modules. These include standard library modules such as ``sys`` and ``math``, which are used for interaction with the interpreter and performing mathematical calculations. A key module is ``rospy``, which is essential for the Robot Operating System (ROS) and later enables communication with the robot. Finally, we use the ``moveit_commander`` module, which provides a user-friendly interface for motion planning via MoveIt.

```python
##### DO NOT CHANGE #####

import sys
import math

import rospy
from moveit_commander import MoveGroupCommander, roscpp_initialize, roscpp_shutdown

##### DO NOT CHANGE #####
```

Now, we initialize the ``moveit_commander`` using the ``roscpp_initialize()`` method and the ROS node using the ``init_node()`` method. For a more detailed explanation of what ``init_node()`` or the argument means, take a look here: https://docs.ros.org/en/melodic/api/rospy/html/rospy-module.html.

```python
##### DO NOT CHANGE #####

roscpp_initialize(sys.argv)
rospy.init_node('exercise_1_1', anonymous=True)

##### DO NOT CHANGE #####
```

Next, we create an instance of the ``MoveGroupCommander()``, setting the ``planner ID`` to the 'RRTConnectkConfigDefault' motion planning algorithm and the ``number of planning attempts`` to 10.

```python
##### DO NOT CHANGE #####

group = MoveGroupCommander("panda_arm")
group.set_planner_id("RRTConnectkConfigDefault")
group.set_num_planning_attempts(10)

##### DO NOT CHANGE #####
```

Now, adjust the 5th axis to a rotation of 155 degrees. Use the ``get_current_joint_values()`` of the ``MoveGroupCommander`` and modify the correct index.

tip:
- use ``math.pi`` 
- ``degrees = radiant x 180 / pi``

```python
joint_goal = group.get_current_joint_values()
joint_goal[4] = (155 * math.pi) / 180
```

Now, we just need to set our ``joint_goal`` as our target and execute the movement. For this use the ``set_joint_value_target()`` and ``go(wait=True)`` methods of the ``MoveGroupCommander``.

```python
group.set_joint_value_target(joint_goal)
group.go(wait=True)
```

Finally, we want to shut down our ``ROS node`` and the ``moveit_commander``.

```python
##### DO NOT CHANGE #####

rospy.signal_shutdown("Task completed")
roscpp_shutdown()

##### DO NOT CHANGE #####
```
