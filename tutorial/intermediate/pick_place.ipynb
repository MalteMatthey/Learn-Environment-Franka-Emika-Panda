```python
#!/usr/bin/env python

import rospy
import moveit_commander
from moveit_commander import PlanningSceneInterface, MoveGroupCommander
from moveit_msgs.msg import CollisionObject
from geometry_msgs.msg import Pose
from shape_msgs.msg import SolidPrimitive
from trajectory_msgs.msg import JointTrajectory, JointTrajectoryPoint
import tf
import math
import sys


def add_collision_objects(planning_scene_interface):
    """Add collision objects to the planning scene."""
    collision_objects = []

    # Table 1
    table1 = CollisionObject()
    table1.id = "table1"
    table1.header.frame_id = "panda_link0"

    primitive1 = SolidPrimitive()
    primitive1.type = SolidPrimitive.BOX
    primitive1.dimensions = [0.2, 0.4, 0.4]

    table1_pose = Pose()
    table1_pose.position.x = 0.5
    table1_pose.position.y = 0.0
    table1_pose.position.z = 0.2
    table1_pose.orientation.w = 1.0

    table1.primitives = [primitive1]
    table1.primitive_poses = [table1_pose]
    table1.operation = CollisionObject.ADD

    collision_objects.append(table1)

    # Table 2
    table2 = CollisionObject()
    table2.id = "table2"
    table2.header.frame_id = "panda_link0"

    primitive2 = SolidPrimitive()
    primitive2.type = SolidPrimitive.BOX
    primitive2.dimensions = [0.4, 0.2, 0.4]

    table2_pose = Pose()
    table2_pose.position.x = 0.0
    table2_pose.position.y = 0.5
    table2_pose.position.z = 0.2
    table2_pose.orientation.w = 1.0

    table2.primitives = [primitive2]
    table2.primitive_poses = [table2_pose]
    table2.operation = CollisionObject.ADD

    collision_objects.append(table2)

    # Object to pick
    obj = CollisionObject()
    obj.id = "object"
    obj.header.frame_id = "panda_link0"

    primitive3 = SolidPrimitive()
    primitive3.type = SolidPrimitive.BOX
    primitive3.dimensions = [0.02, 0.02, 0.2]

    obj_pose = Pose()
    obj_pose.position.x = 0.5
    obj_pose.position.y = 0.0
    obj_pose.position.z = 0.5
    obj_pose.orientation.w = 1.0

    obj.primitives = [primitive3]
    obj.primitive_poses = [obj_pose]
    obj.operation = CollisionObject.ADD

    collision_objects.append(obj)

    # Apply collision objects to the planning scene
    planning_scene_interface.apply_collision_objects(collision_objects)


def main():
    """Main function."""
    moveit_commander.roscpp_initialize(sys.argv)
    rospy.init_node("panda_arm_pick_place", anonymous=True)

    planning_scene_interface = PlanningSceneInterface()
    move_group = MoveGroupCommander("panda_arm")

    # Add collision objects
    add_collision_objects(planning_scene_interface)

    rospy.sleep(2)  # Allow time for objects to load
    rospy.loginfo("Collision objects added to the scene.")

    # Shutdown
    moveit_commander.roscpp_shutdown()


if __name__ == "__main__":
    main()
```
