### Einführung Planungsalgorithmen

#### Was sind Planungsalgorithmen ?

- Methoden, die verwendet werden, um eine Sequenz von Bewegungen oder Aktionen zu berechnen, die Roboter ausführen muss, um Ziel zu erreichen (zB eine Endposition)
- helfen komplexe Probleme, wie überwinden von Hindernissen, präzise Erreichen von Endpunkt zu lösen
- stellen sicher, dass Bewegung effizient, kollisionsfrei und physikalisch möglich

#### Welche Typen von Planungalgorithmen gibt es ?

##### Sampling-basierte Algorithmen

- erstellen zufällige Stichprobe, um Weg von Start zu Ziel zu finden
- Beispiele : RTT, RTTConnect

##### Suchbasierte Algorithmen

- verwenden Raster oder Graphen, um kürzesten Weg zu berechnen
- Beispiel : A*

##### Optimierungsbasierte Algorithmen

- verbessern bestehende Pfade durch Optimierungskriterien, wie Effizienz
- Beispiele : CHOMP, STOMP

Zuerst importieren wir alle wichtigen Module für diese Aufgabe. Rospy und sys kennt ihr ja schon aus vorherigen Aufgaben.
Der MoveGroupCommander ermöglicht die Interaktion mit MoveIt! über die Python API, somit kann man dann den Roboterarm steuern, Bewegungen planen und ausführen.
roscpp_initialize und roscpp_shutdown ermöglichen die Initialisierung, sowie das Beenden der Kommunikation mit ROS von der MoveIt API heraus.
geometry_msgs.msg diehnt dazu, standardisierte Nachrichtenformate bereitzustellen, wie Position(Pose) und Orientierung(Point) des Roboterarms.

```python
##### DO NOT CHANGE #####

import rospy
import sys
from moveit_commander import MoveGroupCommander, roscpp_initialize, roscpp_shutdown
from geometry_msgs.msg import Pose

# Initialize the MoveIt! commander
roscpp_initialize(sys.argv)

# Initialize the ROS node
rospy.init_node('exercise_4_2', anonymous=True)

# Initialize MoveGroupCommander
group = MoveGroupCommander('panda_arm')

##### DO NOT CHANGE #####
```

Als nächstes wollen wir nun den Planungsalgorithmus auswählen, den wir für die Berechnung der Bewegung verwenden wollen. Dazu wird die Methode set_planner_id verwendet, um den spezifischen Planungsalgorithmus auszuwählen.

```python
# Select the desired planning algorithm
group.set_planner_id("RRTConnectkConfigDefault")
```

Nun wir im folgenden die Zielposition des Endeffektors festgelegt. Der Endeffektor ist der Teil eines Roboterarms, der am äußersten Ende der Kinematikkette angebracht ist und direkt mit der Umwelt interagiert. Er ist der funktionale „Werkzeughalter“ des Roboters, dessen Aufgabe es ist, spezifische Aktionen auszuführen. Diese Aktionen hängen stark von der Aufgabe des Roboters ab.

Die Klasse Pose stellt eine Datenstruktur bereit, die die räumliche Position und Orientierung eines Objekts im 3D-Raum beschreibt. Die Position enthält die x-, y- und z-Koordinaten, welche im folgenden Code die Zielposition des Endeffektors im kartesischen Raum (in Metern) definiert.
Die Orientation w beschreibt die Ausrichtung, was für die Orientierung im 3D-Raum genutzt wird.

Ein w=cos(θ/2), das kleiner als 1 ist, zeigt an, dass eine Rotation stattfindet.
Der Rotationswinkel θ kann über arccos(w)⋅2 berechnet werden.

w sollte immer im Bereich [−1,1] liegen, da es mit cos(θ/2) zusammenhängt. Werte außerhalb dieses Bereichs führen zu ungültigen Quaternionen.

TODO : Orientierung besteht ein aus Quarternion x,y,z,w -> soll man darauf eingehen ?

```python
target_pose = Pose()
target_pose.position.x = 0.4
target_pose.position.y = 0.1
target_pose.position.z = 0.4
target_pose.orientation.w = 1.0
```

Jetzt geht es darum, die Bewegung des Roboterarms zu planen und auszuführen.

Die Methode group.go() führt die geplante Bewegung des Roboters aus, basierend auf der Zielpose, die zuvor festgelegt wurde.

Das Argument wait=True bedeutet, dass die weitere Ausführung des Codes blockiert wird(d.h. wartet), bis die Bewegung abgeschlossen ist. Erst nach Abschluss der Bewegung wird der nächste Befehl ausgeführt.

Die Methode group.stop() stoppt die Bewegung des Roboters sofort.
Auch wenn die Bewegung bereits abgeschlossen ist, wird mit diesem Befehl sichergestellt, dass keine weiteren Bewegungen oder Restbewegungen (z. B. durch Trägheit oder Nachschwingen) stattfinden.

Die Methode group.clear_pose_targets() entfernt alle Zielpositionen und -orientierungen (Pose Targets), die zuvor für den Roboter definiert wurden.

Dies ist sehr wichtig, da Zielpositionen, die nicht entfernt werden, bleiben im Speicher des MoveIt-Objekts (group) und könnten unbeabsichtigt bei späteren Bewegungsbefehlen wiederverwendet werden.

```python
##### DO NOT CHANGE #####

plan = group.go(wait=True)
group.stop()
group.clear_pose_targets()

if plan:
    rospy.loginfo("Bewegung erfolgreich geplant und ausgeführt!")
else:
    rospy.logwarn("Planung fehlgeschlagen. Überprüfe die Konfiguration.")

##### DO NOT CHANGE #####
```
