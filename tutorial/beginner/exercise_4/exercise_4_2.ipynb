Zuerst importieren wir alle wichtigen Module für diese Aufgabe. Rospy und sys kennt ihr ja schon aus vorherigen Aufgaben.
Der MoveGroupCommander ermöglicht die Interaktion mit MoveIt! über die Python API, somit kann man dann den Roboterarm steuern, Bewegungen planen und ausführen.
roscpp_initialize und roscpp_shutdown ermöglichen die Initialisierung, sowie das Beenden der Kommunikation mit ROS von der MoveIt API heraus.

```python
##### DO NOT CHANGE #####

import rospy
import sys
from moveit_commander import MoveGroupCommander, roscpp_initialize, roscpp_shutdown

# Initialize the MoveIt! commander
roscpp_initialize(sys.argv)

# Initialize the ROS node
rospy.init_node('exercise_4_2', anonymous=True)

##### DO NOT CHANGE #####
```

Nun wird die MoveGroup für den Roboterarm `panda_arm` initialisiert.
Zusätzlich werden Planungsparameter wie der Planungsalgorithmus und die Anzahl der Planungsversuche festgelegt.


```python
##### DO NOT CHANGE #####

# Initialize the MoveGroup
group = MoveGroupCommander('panda_arm')

# Set the planner ID to 'RRTConnectkConfigDefault'
group.set_planner_id("RRTConnectkConfigDefault")

# Set the number of planning attempts to 10
group.set_num_planning_attempts(10)

##### DO NOT CHANGE #####
```

### Abfrage der aktuellen Gelenkwinkel
Die aktuellen Gelenkwinkel des Roboterarms werden abgefragt und im Log ausgegeben.


```python
# Query the current joint values
joint_values = group.get_current_joint_values()
rospy.loginfo("Current Joint Values: ")
for i, joint in enumerate(joint_values):
    rospy.loginfo(f"Joint {i+1}: {joint:.2f} rad")
```

### Abfrage der aktuellen Endeffektor-Position
Die aktuelle Position und Orientierung des Endeffektors werden abgefragt und im Log ausgegeben.


```python
# Query the current end-effector pose
pose = group.get_current_pose().pose
rospy.loginfo(f"Endeffector Position: x={pose.position.x:.3f}, y={pose.position.y:.3f}, z={pose.position.z:.3f}")
rospy.loginfo(f"Endeffector Orientation: x={pose.orientation.x:.3f}, y={pose.orientation.y:.3f}, z={pose.orientation.z:.3f}, w={pose.orientation.w:.3f}")
```

### Aufräumen und Beenden des ROS-Knotens
Am Ende wird der ROS-Knoten heruntergefahren und die Ressourcen werden freigegeben.


```python
##### DO NOT CHANGE #####

# Shutdown the ROS node
rospy.signal_shutdown("Task completed")
roscpp_shutdown()

##### DO NOT CHANGE #####
```
