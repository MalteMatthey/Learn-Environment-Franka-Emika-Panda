```python
#!/usr/bin/env python

import rospy
from control_msgs.msg import FollowJointTrajectoryAction, FollowJointTrajectoryGoal
from trajectory_msgs.msg import JointTrajectoryPoint
import actionlib

# ROS-Knoten initialisieren
rospy.init_node("position_controller_example", anonymous=True)

# Verbindung mit dem Trajectory Controller herstellen
# Der Code verwendet den ROS-Action-Server /panda_arm_controller/follow_joint_trajectory, um Gelenkpositionen anzusteuern.
# Diese Schnittstelle basiert auf dem Trajectory Controller, der für die präzise Steuerung von Gelenkbewegungen zuständig ist.

client = actionlib.SimpleActionClient('/panda_arm_controller/follow_joint_trajectory', FollowJointTrajectoryAction)
rospy.loginfo("Waiting for trajectory controller...")
client.wait_for_server()
rospy.loginfo("Connected to trajectory controller!")

# Zielgelenkpositionen definieren
joint_goal = [0.0, -0.785, 0.0, -2.356, 0.0, 1.57, 0.785]

# Trajektorie erstellen
goal = FollowJointTrajectoryGoal()
goal.trajectory.joint_names = [
    "panda_joint1", "panda_joint2", "panda_joint3",
    "panda_joint4", "panda_joint5", "panda_joint6", "panda_joint7"
]

point = JointTrajectoryPoint()
point.positions = joint_goal
point.time_from_start = rospy.Duration(5.0)  # Die Bewegung soll in 5 Sekunden abgeschlossen sein

goal.trajectory.points.append(point)
goal.trajectory.header.stamp = rospy.Time.now()

# Ziel senden
rospy.loginfo("Sending goal to trajectory controller...")
client.send_goal(goal)
client.wait_for_result()

# Ergebnis überprüfen
result = client.get_result()
if result:
    rospy.loginfo("Trajectory successfully executed!")
else:
    rospy.logwarn("Failed to execute trajectory.")
```
