We start by creating a ``JointTrajectory()`` object, the ROS message type that will later be used to publish our target joint values. Before that, we need to define its ``header`` and specify the ``joint_names`` to the default names.

The default joint values are [0, -45, 0, -135, 0, 90, 45] degrees.

```python
if decision == 'move_group':
    ##### YOUR CODE HERE IF YOU CHOOSED MOVEGROUPCOMMANDER #####

    roscpp_initialize(sys.argv)

    group = MoveGroupCommander("panda_arm")
    group.set_planner_id("RRTConnectkConfigDefault")
    group.set_planning_time(10)

    group.set_max_velocity_scaling_factor(1.0)  # Set to 1.0 for maximum speed
    group.set_max_acceleration_scaling_factor(1.0)

    joint_goal = group.get_current_joint_values()

    joint_goal[0] = 0.0
    joint_goal[1] = -math.pi / 4
    joint_goal[2] = 0.0
    joint_goal[3] = -math.pi * 3 / 4
    joint_goal[4] = 0.0
    joint_goal[5] = math.pi / 2
    joint_goal[6] = math.pi / 4

    group.set_joint_value_target(joint_goal)
    group.go(wait=True)

    roscpp_shutdown()

    ##### YOUR CODE HERE IF YOU CHOOSED MOVEGROUPCOMMANDER #####

elif decision == 'rostopic':
    ##### YOUR CODE HERE IF YOU CHOOSED MOVEGROUPCOMMANDER #####

    pub = rospy.Publisher('/effort_joint_trajectory_controller/command', JointTrajectory, queue_size=10)

    joint_trajectory = JointTrajectory()
    joint_trajectory.header.stamp = rospy.Time.now()
    joint_trajectory.joint_names = ['panda_joint1', 'panda_joint2', 'panda_joint3', 'panda_joint4', 'panda_joint5', 'panda_joint6', 'panda_joint7']

    current_joint_values = rospy.wait_for_message('/joint_states', JointState, timeout=5).position
    point = JointTrajectoryPoint()
    point.positions = list(current_joint_values)

    point.positions[0] = (0 * math.pi) / 180
    point.positions[1] = -(45 * math.pi) / 180
    point.positions[2] = (0 * math.pi) / 180
    point.positions[3] = -(135 * math.pi) / 180
    point.positions[4] = (0 * math.pi) / 180
    point.positions[5] = (90 * math.pi) / 180
    point.positions[6] = (45 * math.pi) / 180

    point.time_from_start = rospy.Duration(2.5) 

    joint_trajectory.points.append(point)
    rospy.sleep(1.0)
    pub.publish(joint_trajectory)

    rospy.sleep(2.5)
    ##### YOUR CODE HERE IF YOU CHOOSED MOVEGROUPCOMMANDER #####

else:
    raise NotImplementedError
```
