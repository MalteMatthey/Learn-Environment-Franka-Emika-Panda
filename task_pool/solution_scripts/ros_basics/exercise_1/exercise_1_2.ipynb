First we need to import the Modules.

```python
##### DO NOT CHANGE #####

import rospy
import time

from std_msgs.msg import String

##### DO NOT CHANGE #####
```

In the background, we will have a ``ROS node`` that ``publishes`` String messages on the ``/notification`` ``topic``. First, we need to initialize the ``callback()`` method of the ``Subscriber``. If the ``Subscriber`` receives any messages on the subscribed ``topic``, it will call a method that we specify, passing the received message as an argument to this method. We want our callback method to update the global variable ``received_msg`` with the newest message received by the subscriber.

```python
##### DO NOT CHANGE #####

def callback(msg):

    global received_msg
    received_msg = msg.data
    rospy.loginfo("set to: " + received_msg)

##### DO NOT CHANGE #####
```

Now, we will specify the ``Subscriber``. First, we initialize a ``ROS node``. Then, we set the ``topic`` that the subscriber should listen to as ``/notification``. We also define the data type of the received messages as ``String`` and specify the method that the subscriber should call, passing the received data as an argument. The ``spin()`` method from ``rospy`` keeps the node running and processing incoming messages.

ToDo:
- initialize the ``ROS node`` with the name ``'subscriber_node'``

```python
##### YOUR CODE HERE #####

def subscribe():

    rospy.init_node("subscriber_node", anonymous=True) # implement by users

    rospy.sleep(2)

    rospy.loginfo("subscriber_node is running")

    rospy.Subscriber("/notification", String, callback)

    rospy.loginfo("subscriber_node has started")

    rospy.spin()

##### YOUR CODE HERE #####
```

Now we just need to start the ``subscribe()`` method.

```python
##### DO NOT CHANGE #####

if __name__ == '__main__':
    received_msg = None
    subscribe()

##### DO NOT CHANGE #####
```
